```python
from chalk.client import ChalkClient
from chalk import _
from src.models import User, Transaction
```

```python
client = ChalkClient(branch="main")
```

## Online queries from code












```python
client.query(
    input={User.id: 1},
    output=[
        User,
        User.credit_report.percent_past_due,
    ],
    staleness={
        User.credit_report.percent_past_due: "0s",
    }
)
```

## Offline queries


```python
client.offline_query(
    output=[
        User.name,
        User.denylisted,
        User.name_email_match_score,
        User.domain_name,
        User.email_username,
        User.total_spend,
        User.credit_report.percent_past_due,
    ],
    recompute_features=[User.credit_report.percent_past_due],
    max_samples=30,
).to_pandas()
```

```python
client.offline_query(
    max_samples=100,
    output=[
        Transaction.memo,
        Transaction.user_id,
        Transaction.amount
    ],
    recompute_features=True
).to_pandas()
```
